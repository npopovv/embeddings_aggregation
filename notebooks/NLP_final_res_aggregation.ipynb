```python
def save_latex(df, path, caption, label, float_format="%.3f"):
    df_formatted = df.copy()

    for row in df_formatted.index:
        max_val = df_formatted.loc[row].max()
        for col in df_formatted.columns:
            val = df_formatted.loc[row, col]
            formatted = float_format % val
            formatted = formatted.rstrip('0').rstrip('.') if '.' in formatted else formatted
            if val == max_val:
                formatted = r"\textbf{" + formatted + r"}"
            df_formatted.loc[row, col] = formatted

    df_formatted = df_formatted.astype(str)

    col_format = 'l' + 'l' * df.shape[1]  # используем 'l', чтобы LaTeX не обрабатывал как числа

    with open(path, "w", encoding="utf8") as f:
        f.write(r"\begin{table}[H]" + "\n")
        f.write(r"\centering" + "\n")
        f.write(r"\small" + "\n")
        f.write(r"\renewcommand{\arraystretch}{1.2}" + "\n")
        f.write(r"\resizebox{\textwidth}{!}{" + "\n")
        f.write(df_formatted.to_latex(index=True, escape=False, column_format=col_format))
        f.write("}\n")
        f.write(r"\caption{" + caption + "}" + "\n")
        f.write(r"\label{" + label + "}" + "\n")
        f.write(r"\end{table}" + "\n")



save_latex(bert, "bert_results.tex", "Результаты для модели BERT", "tab:bert_results")
save_latex(roberta, "roberta_results.tex", "Результаты для модели RoBERTa", "tab:roberta_results")
save_latex(simcse, "simcse_results.tex", "Результаты для модели SimCSE", "tab:simcse_results")
```

### Средний ранг

```python
# Загрузи таблицы (с уже переименованными колонками)
roberta = pd.read_csv(PATH+'ROBERTA_final_results_1.csv', index_col = 0)
bert = pd.read_csv(PATH+'BERT_final_results_1.csv', index_col = 0)
simcse = pd.read_csv(PATH+'SIMSCE_final_results_1.csv', index_col = 0)

method_names = {
    "last_mean": "Last Layer",
    "best_layer": "Best Layer",
    "weighted_mean_qp": "QP Weighted",
    "concat_pca_cluster": "Cluster + PCA",
    "concat_pca_qp": "QP + PCA",
    "greedy": "Greedy",
    "qp_nonsym": "QP masked"
}
bert.rename(columns=method_names, inplace=True)
roberta.rename(columns=method_names, inplace=True)
simcse.rename(columns=method_names, inplace=True)
```

```python
# Список DataFrame'ов по моделям (вручную или загруженные из файлов)
dfs = [bert, roberta, simcse]

# Приводим имена методов к единому списку (в simcse 'Pooling' — это 'Last Layer')
rename_map = {'Pooling': 'Last Layer'}
for df in dfs:
    df.rename(columns=rename_map, inplace=True)

# Вычисляем ранги (меньше — лучше)
ranks = []
for df in dfs:
    ranked = df.rank(axis=1, ascending=False, method='min')  # одинаковые значения получают одинаковый наименьший ранг
    ranks.append(ranked)

# Суммируем и усредняем ранги
total_ranks = sum(ranks)/3
average_ranks = total_ranks.mean(axis=0).sort_values()

# Печать результатов
print("Средний ранг методов:")
print(average_ranks)
```

```python

# Построение графика
plt.figure(figsize=(10, 6))
plt.barh(average_ranks.index, average_ranks, color='skyblue')
plt.xlabel('Средний ранг (меньше — лучше)')
plt.title('Сравнение методов объединения по среднему рангу')
plt.gca().invert_yaxis()  # Лучшие вверху
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig("method_ranking.png", dpi=300, bbox_inches='tight')  # путь и настройки можно изменить
plt.show()
```
